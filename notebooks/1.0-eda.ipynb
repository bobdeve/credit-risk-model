```python
import pandas as pd

# Load the dataset
df = pd.read_csv('../data/data.csv')  # adjust the path as needed

```

```python
# Quick look at structure
print("Shape of the dataset:", df.shape)

# Show data types and missing values
df.info()

# View first few rows
df.head()
```

```python
df.describe(include='all').transpose()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['FraudResult'].value_counts(normalize=True)
```

```python
df['TransactionStartTime'] = pd.to_datetime(df['TransactionStartTime'])

# Extract useful time features
df['Hour'] = df['TransactionStartTime'].dt.hour
df['DayOfWeek'] = df['TransactionStartTime'].dt.dayofweek
```

```python
df.corr(numeric_only=True)['FraudResult'].sort_values(ascending=False)
```

```python
# Monetary = total value per customer
monetary = df.groupby('CustomerId')['Value'].sum().reset_index()

# Frequency = total transactions per customer
frequency = df.groupby('CustomerId')['TransactionId'].count().reset_index()

# Recency = days since last transaction
recent_time = df['TransactionStartTime'].max()
recency = df.groupby('CustomerId')['TransactionStartTime'].max().reset_index()
recency['Recency'] = (recent_time - recency['TransactionStartTime']).dt.days
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numeric columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Plot histograms for all numeric columns
df[numeric_cols].hist(figsize=(15, 10), bins=30)
plt.suptitle('Distribution of Numerical Features', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Plot bar charts for top categorical features
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    plt.figure(figsize=(10, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().iloc[:10].index)
    plt.xticks(rotation=45)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Heatmap for correlation
plt.figure(figsize=(12, 8))
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Check missing values
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
missing_percent = (missing / len(df)) * 100

# Display missing data
missing_data = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_percent})
missing_data
```

```python
# Boxplots to detect outliers in important numeric features
for col in ['Amount', 'Value']:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
# Skewness of numeric features
df[numeric_cols].skew().sort_values(ascending=False)
```
